```python
disarium_numbers = []

for num in range(1, 10001):
    digits = str(num)
    total = sum(int(digit) ** (idx + 1) for idx, digit in enumerate(digits))
    if total == num:
        disarium_numbers.append(num) 


print("The numbers which are disarium  numbers are as follows:- ", disarium_numbers) 
```
